```python
import requests 
import bs4 
import pandas as pd
```

```python
sample = [] #List of Dictionary{"google_reasult","youtube_link","youtube_channel_link"}


for results in range(0,10000,10):
  keyword = "site:youtube.com+openinapp.co"
  # Generating the url
  url = "https://google.com/search?q=" + keyword + "&ie=UTF-8&ei=1rd1ZNqlHM-3qtsP0Oyl8AQ&start=" +str(results)+ "&sa=N"

  # print(url)


  request_result = requests.get(url)

  soup = bs4.BeautifulSoup( request_result.text  , "html.parser" )

  temp = soup.find_all('div', {'class':'egMi0 kCrYT'})


  # print(temp[2].a['href'][7:])
  #///url?

  for result in temp:
    print(result)
    # print(result.a['href'][7:])
    sample.append({"google_result":"https://www.google.com/url?sa=t&source=web&rct=j&url=" + result.a['href'][7:]})
    # print("https://www.google.com/url?sa=t&source=web&rct=j&url=" + result.a['href'][7:])

#Sample now has all google result links
```

```python
len(sample)
```

```python
for i in sample:
  suburl = i['google_result']
  sub_request_result = requests.get( suburl )
  sub_soup = bs4.BeautifulSoup( sub_request_result.text  , "html.parser" )
  youtube_link = sub_soup.find('meta')['content'].split("url=")[1]
  i['youtube_link'] = youtube_link

  sub_request_result = requests.get( youtube_link )
  sub_soup = bs4.BeautifulSoup( sub_request_result.text  , "html.parser" )
  try:
    youtube_channel_link = sub_soup.find('span', {"itemprop":"author"}).link['href']
  except:
    youtube_channel_link = 'No link'
  i['youtube_channel_link'] = youtube_channel_link
```

```python
df = pd.DataFrame.from_records(sample)
```
